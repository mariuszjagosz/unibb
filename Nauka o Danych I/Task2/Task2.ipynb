```python
import pandas as pd
file_path = 'IHME_DAH_DATABASE_1990_2020_Y2021M09D22.CSV'
chunksize = 100000 
df_csv = pd.concat(pd.read_csv(file_path, chunksize=chunksize, encoding="utf-8", low_memory=False))
```

```python
df_csv.head(10)
```

```python
cols_to_convert = df_csv.columns[14:]
df_csv[cols_to_convert] = df_csv[cols_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
import numpy as np
```

```python
mean_val = np.mean(df_csv["dah_20"])
```

```python
print(mean_val)
```

```python
df_csv.dropna(subset=["dah_20"], inplace=True)
median_val = np.median(df_csv["dah_20"])
print(median_val)
```

```python
std_dev = np.std(df_csv["dah_20"])
print(std_dev)
```

```python
variance = np.var(df_csv["dah_20"])
print(variance)
```

```python
df_csv.dropna(subset=["other_dah_20"], inplace=True)
data_1 = df_csv["other_dah_20"]
data_2 = df_csv["dah_20"]
correlation = np.corrcoef(data_1, data_2)[0, 1]
print(correlation)
```

```python
covariance = np.cov(data_1, data_2)[0, 1]
print(covariance)
```
