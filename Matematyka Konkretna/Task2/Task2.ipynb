```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('war3.csv', sep=';')
data['y'] = data['y'].str.replace(',', '.').astype(float)

A = data[['x1', 'x2']].values
b = data['y'].values

U, S, VT = np.linalg.svd(A, full_matrices=False)

S_inv = np.diag(1 / S)
x = VT.T @ S_inv @ U.T @ b

y_pred = A @ x

print("Współczynniki regresji [a, b]:", x)

plt.figure(figsize=(10, 5))
plt.plot(b, 'ko-', label='Dane rzeczywiste (y)')
plt.plot(y_pred, 'ro--', label='Regresja (ŷ)')
plt.title('Regresja wielowymiarowa – porównanie wartości')
plt.xlabel('Próbka')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python

x_final = np.linalg.pinv(A) @ b
y_pred_final = A @ x_final

plt.figure(figsize=(10, 5))
plt.plot(b, 'ko-', label='Dane rzeczywiste (y)')
plt.plot(y_pred_final, 'ro--', label='Regresja (ŷ) z pinv')
plt.title('Regresja wielowymiarowa – wynik przy użyciu np.linalg.pinv')
plt.xlabel('Próbka')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

x_final

```
