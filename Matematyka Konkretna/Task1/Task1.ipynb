```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from skimage.color import rgb2gray

img = Image.open("3.webp").convert("RGB")
img_gray = rgb2gray(np.array(img))
plt.imshow(img_gray, cmap='gray')
plt.title("Obraz w skali szaroci")
plt.axis("off")
plt.show()

U, S, VT = np.linalg.svd(img_gray, full_matrices=False)
Sigma = np.diag(S)

X = img_gray
correlation_rows = X @ X.T    # XX^T
correlation_cols = X.T @ X    # X^TX

print('\n')
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_rows, cmap='viridis')
plt.title("Macierz korelacji wierszy (XX)")
plt.xlabel("Wiersze")
plt.ylabel("Wiersze")
plt.show()
print('\n')
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_cols, cmap='viridis')
plt.title("Macierz korelacji kolumn (XX)")
plt.xlabel("Kolumny")
plt.ylabel("Kolumny")
plt.show()
print('\n')
r = 50
X_approx = U[:, :r] @ Sigma[:r, :r] @ VT[:r, :]
plt.imshow(X_approx, cmap='gray')
plt.title(f"Przyblienie obrazu przez {r} skadowych SVD")
plt.axis("off")
plt.show()
```
