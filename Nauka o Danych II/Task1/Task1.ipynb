Zadanie 1. Wariant zadania numer 2 - nie da się zamodelować funkcji cos(x)**sin(x) w przedziale 0 do 2 * pi, ponieważ, wtedy dostajemy Nan (podnosimy ujemne wartości do ułakowej potęgi dostając liczby zespolone), zmieniono więc funkcję na cos(x)*sin(x)

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_squared_error

np.random.seed(42)
X = np.linspace(0,2* np.pi, 1000)
y = np.cos(X) * np.sin(X)

model = Sequential([
    Dense(256, activation='relu', input_shape=(1,)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=100, batch_size=32, verbose=1)

y_pred = model.predict(X)

mse = mean_squared_error(y, y_pred)
print(f"Ostateczny błąd MSE na całym zbiorze: {mse:.5f}")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(X, y, label='Funkcja oryginalna', color='blue')
plt.scatter(X, y_pred, color='red', alpha=0.5, s=10, label='Predykcja sieci neuronowej')
plt.xlabel('X (0 do 2π)')
plt.ylabel('Wartości funkcji')
plt.title('Modelowanie funkcji  za pomocą sieci neuronowej')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_squared_error

np.random.seed(42)
X = np.random.rand(1000, 4) * [200, 10, 3, 1]  # powierzchnia, mieszkańcy, ogrzewanie, izolacja
y = np.random.rand(1000, 3) * [5000, 2000, 25]  # energia elektryczna, gaz, temperatura

model = Sequential([
    Dense(128, activation='relu', input_shape=(4,)),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(3)
])

model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=300, batch_size=32, verbose=1)

y_pred = model.predict(X)

mse = mean_squared_error(y, y_pred)
print(f"Ostateczny błąd MSE na całym zbiorze: {mse:.5f}")
```
